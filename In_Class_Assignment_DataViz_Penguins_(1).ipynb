## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()
```

**Your interpretation (2–3 sentences):**

Adelie penguins have shorter bills but deeper ones compared to other species. Chinstrap penguins are in the middle range among all species. Gentoo penguins have almost the same bill length as Chinstrap penguins, but their bills are thinner. Chinstrap penguins have long and thick bills.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

plt.hist(penguins['bill_length_mm'], bins=10)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# TODO: Histogram with 30 bins
plt.hist(penguins['bill_length_mm'], bins=30)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

**Your comparison (1–2 sentences):**

Compared to 10 bins, 30 bins show more detailed characteristics of the data. However, using 10 bins provides a rough idea of which length appears most frequently.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

**Your summary (1–2 sentences):**

Gentoo penguins have a much higher body mass than Adelie and Chinstrap penguins, while Adelie and Chinstrap show similar median values with a few outliers.
